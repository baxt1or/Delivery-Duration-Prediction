```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
delivery_df["created_at"] = pd.to_datetime(delivery_df["created_at"])
delivery_df["actual_delivery_time"] = pd.to_datetime(delivery_df["actual_delivery_time"])

delivery_df["actual_delivery_duration"] = (delivery_df["actual_delivery_time"] - delivery_df["created_at"]).dt.total_seconds()


delivery_df.tail()
```

```python
delivery_df = delivery_df.fillna(0)
```

```python
to_drop = ["created_at","market_id" ,"actual_delivery_time", "store_primary_category", "actual_delivery_duration"]

X = delivery_df.drop(to_drop, axis=1)
y = delivery_df["actual_delivery_duration"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.pipeline import Pipeline


def model_building(X, y, model, scaler):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    model_pipeline = Pipeline([
        ("scaler", scaler), 
        ("model", model)
    ])
    
    model_pipeline.fit(X_train, y_train)

    y_pred = model_pipeline.predict(X_test)

    print("R2 Score: ", r2_score(y_test, y_pred))
    print("MSE: ", mean_squared_error(y_test, y_pred))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, MinMaxScaler

lr = LinearRegression()
scaler = StandardScaler()

model_building(X, y, lr, scaler)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

model_building(X, y, model, scaler)
```
